```python
import pandas as pd
import sys
sys.path.append('./FX-1-Minute-Data-master/full-data-2000-2017/')
```

```python
from zipfile import ZipFile
import glob
```

```python
def load_data()
    list_paths = glob.glob('*/full-data-2000-2017/**/', recursive=True)
    list_years = glob.glob('*/full-data-2000-2017/**/*.zip', recursive=True)

    df = pd.DataFrame()
    for path in list_paths[1:]:
        list_years = glob.glob(path+'*.zip', recursive=True)

        df_tmp = pd.DataFrame()
        for year in list_years:
            ticker = year[len("FX-1-Minute-Data-master/full-data-2000-2017/eurczk/DAT_ASCII_"):][:len("EURCZK")]
            filename = year[len("FX-1-Minute-Data-master/full-data-2000-2017/eurczk/"):-len('.zip')]

            columns = ['DateTime Stamp', ticker + ' CLOSE Bid Quote']
            zip_file = ZipFile(year)
            df_tmp = pd.concat([df_tmp, pd.read_csv(zip_file.open(filename + '.csv'), sep=';', index_col=[0], usecols=[0, 4], names=columns)], axis=0)

        df = pd.concat([df, df_tmp], axis=1)
        
    return df
```
